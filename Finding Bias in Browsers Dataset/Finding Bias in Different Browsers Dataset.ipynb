```python
from deepface import DeepFace
import cv2
import os
import matplotlib.pyplot as plt
```

# New Section

```python
from google.colab import drive, files
drive.mount('/content/drive')
path = "/content/drive/My Drive/Colab Notebooks/ML Project/"
```

```python
browsers = ['Bing', 'DuckDuckGo', 'Google']
nationalities = ['Chinese', 'French', 'Italian', 'Japanese', 'Kenyan', 'Korean', 'Pakistani', 'Russian', 'Thai', 'Turkish']

for x in browsers:
  for j in nationalities:
    path_per_nation = path + x +'/'+ j
    file_names  = os.listdir(path_per_nation)
    final =[]

    for i in range(0,len(file_names)):
      img = cv2.imread(str(path_per_nation+'/'+file_names[i]))
      #plt.imshow(img[:,:,:])
      #plt.show()
      
      result = DeepFace.analyze(img,actions = ['gender'], enforce_detection=False)
      final.append(result['gender'])
      print("Gender is: ", final[i])
      name = str(final[i])+'_'+str(i)
      #os.rename(str(path_per_nation)+'/'+str(file_names[i]),name)

    male_count = final.count("Man")
    female_count =  final.count('Woman')

    
    # Creating dataset
    gender = ['Man', 'Woman']
    
    data = [male_count, female_count]
    
    # Creating plot
    fig = plt.figure(figsize =(10, 7))
    plt.title(str(j)+ ' Data ' +str(x))
    plt.pie(data, labels = gender, shadow =True, autopct = '%1.1f%%')
    plt.legend(gender,
              title ="Gender")
    save_name = str(j)+'Data'+str(x)
    plt.savefig(save_name)
    # show plot
    plt.show()
```
